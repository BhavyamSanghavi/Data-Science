```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv('Iris.csv')
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
dfg=df.groupby('Species')
dfg.describe()
```

```python
columns = ["SepalLengthCm", "SepalWidthCm", "PetalLengthCm", "PetalWidthCm"]
print(columns)
mean=[]
median=[]
var=[]

# Group the DataFrame by species
df_grouped = df.groupby('Species')

for species, group in df_grouped:
    print(f"Statistics for {species}:\n")
    
    for col in columns:
        val = group[col]  # Get values for the current column of the species
        
        # Calculate variance, standard deviation, mean, median, etc.
        variance = sum((val - val.mean())**2) / len(val)
        deviation = variance**0.5
        
        print(f"Column: {col}")
        print(f"Variance = {variance}")
        print(f"Standard Deviation = {deviation}")
        print(f"Mean = {val.mean()}")
        print(f"Median = {val.median()}")
        print(f"Inter Quartile Range = {val.quantile(0.75) - val.quantile(0.25)}")
        print(f"Min = {val.min()}")
        print(f"Max = {val.max()}")
        print("\n")
        mean.append(val.mean())
        median.append(val.median())
        var.append(variance)
```

### NBA Dataset

```python
df=pd.read_csv("nba.csv")

# Data Cleaning

df["College"]=df["College"].fillna("No College")
df["Salary"]=df["Salary"].fillna(df["Salary"].mean())
df=df.dropna()
df.isna().sum()
```

```python
row = df['Age']
for i in range(len(row)):
    if row[i] < 20:
        row[i] = "Teen"
    elif row[i] < 30:
        row[i] = "Young Adult"
    elif row[i] < 40:
        row[i] = "Adult"
    else:
        row[i] = "Old"

# Convert the 'Age' column to object type before grouping
df['Age'] = df['Age'].astype('object')
dfg = df.groupby('Age')

```

```python
dfg.describe()
```

```python
columns = ["Salary"]
print(columns)

for ageGroup, group in dfg:
    print(f"Statistics for {ageGroup}:\n")
    
    for col in columns:
        val = group[col]  # Get values for the current column of the species
        
        # Calculate variance, standard deviation, mean, median, etc.
        variance = sum((val - val.mean())**2) / len(val)
        deviation = variance**0.5
        
        print(f"Column: {col}")
        print(f"Variance = {variance}")
        print(f"Standard Deviation = {deviation}")
        print(f"Mean = {val.mean()}")
        print(f"Median = {val.median()}")
        print(f"Inter Quartile Range = {val.quantile(0.75) - val.quantile(0.25)}")
        print(f"Min = {val.min()}")
        print(f"Max = {val.max()}")
        print("\n")
```
