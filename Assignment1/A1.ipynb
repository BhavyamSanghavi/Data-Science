```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import LabelEncoder
```

```python
df=pd.read_csv("nba.csv")
df.head()
```

```python
df.describe()
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df['College'].fillna('No College', inplace=True)        ## replace null collge by string 'No College'
df['Salary'].fillna(df['Salary'].mean(), inplace=True)  ## replace null salaries with average salaries
df.dropna(inplace=True)         ## drop remaining null values
df
```

```python
df.isna().sum()
```

```python
# to get distinct values from column
unique_values = df['Team'].unique()
len(unique_values)

uq=df['Position'].unique()
uq
```

```python
# one hot encoding -->  Adds additional columns to the data
dfp=pd.get_dummies(df,columns=['Position'])
dfp
```

```python
# label encoding -->  don't add additional columns and replace values from 0-n-1 inplace
df['Position']=LabelEncoder().fit_transform(df['Position'])
df.head(20)
```

```python
df.info()
```

```python
df["Age"]=df["Age"].astype(int)
df["Name"]=df["Name"].astype(str)
df.info()
```
