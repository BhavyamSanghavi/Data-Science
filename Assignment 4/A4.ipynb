```python
import pandas as pd
import seaborn as sb
```

```python
df=pd.read_csv('Titanic.csv')
df.head()
```

```python
df.isnull().sum()
```

```python
df['Age'].fillna(df['Age'].mean(),inplace=True)
df['Cabin'].fillna('Unknown',inplace=True) 
df['Embarked'].fillna('Unknown',inplace=True)
# fill the values of fare where fare is zero with the mean of the fare
mean=75
print(mean)
row=df['Fare']
for i in range(len(df['Fare'])):
    if row[i]<30:
        print("Replacing")
        row[i]=mean
# df.isnull().sum()
```
